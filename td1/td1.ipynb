## EXERCICE 18

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
base = {"y" : [2.1, 2.6, 3.1, 2.8, 3, 2.7, 3.4, 3.5, 5, 4.9, 4.8, 4.6],
        "x" : [3.4, 3.6, 4.5, 4.3, 4.7, 4, 4.8, 5.1, 7, 7.2, 7.1, 6.9]}

base = pd.DataFrame(base)

x = base["x"]
x = sm.add_constant(x)

y = base["y"]
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x = base["x"], y = base["y"])
```

```python
model = sm.OLS(y, x).fit()
print_model = model.summary()
print_model
```

```python
plt.figure(figsize=(10,5))
plt.plot(model.resid, ".")
plt.title("Graphique des résidus estimés")
```

## EXERCICE 22

```python
import pandas as pd
import yfinance as yf
from yahoofinancials import YahooFinancials
```

```python
carrefour_df = yf.download('CA.PA', 
                      start='2019-01-01', 
                      end='2019-12-31', 
                      progress=False)

sanofi_df = yf.download('SAN', 
                      start='2019-01-01', 
                      end='2019-12-31', 
                      progress=False)

carrefour = pd.DataFrame(carrefour)
sanofi = pd.DataFrame(sanofi)
```

```python
base = sanofi.merge(carrefour,left_index=True, right_index=True)
base = base.dropna()
base = base.rename(columns = {"Close_x" : "sanofi", "Close_y" : "carrefour"})
```

```python
plt.figure(figsize=(10,5))
plt.plot(base["carrefour"], label = "Cours carrefour")
plt.plot(base["sanofi"], label="Cours Sanofi")
plt.legend()
plt.title("Evolution des cours dans")
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x = base["carrefour"], y = base["sanofi"])
plt.xlabel("Action Carrefour")
plt.ylabel("Action Sanofi")
plt.title("Diagramme des deux cours")
```

```python
y = base["sanofi"]

x = base["carrefour"]
x = sm.add_constant(x)
```

```python
model = sm.OLS(y, x).fit()
print_model = model.summary()
print_model
```

```python
plt.figure(figsize=(10,5))
plt.plot(model.resid, ".")
plt.title("Graphique des résidus estimés")
```

```python
predictions = model.predict(x) 
plt.figure(figsize=(10,5))
plt.plot(predictions, "--", label = "Prédiction", color="blue")
plt.plot(base["sanofi"], label = "Cours Sanofi", color = "black")
plt.title("Graphique des prédictions")
```
